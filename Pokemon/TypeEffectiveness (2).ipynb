```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
import numpy as np
```

```python
types = ['normal', 'fire', 'water', 'grass', 'electric', 'ice', 'fighting', 'poison', 'ground', 'flying', 'psychic', 'bug', 'rock', 'ghost', 'dark', 'dragon', 'steel', 'fairy']      
types_dict = {}    


for type in types:    
    
    super_effective = []
    not_very_effective = []
    no_effect = []

    url = f'https://pokemondb.net/type/{type}'
    data  = requests.get(url).text
    soup = BeautifulSoup(data,"lxml")
    bm = soup.body.main
    bmd = bm.find('div', class_="grid-col span-md-6 span-lg-12")
    bmdp = bmd.find_all('p', class_="type-fx-list")

    if type != 'normal':
        for a in bmdp[0].find_all('a'):
            super_effective.append(a.text.lower())  
        for a in bmdp[1].find_all('a'):
            not_very_effective.append(a.text.lower())
        try:
            for a in bmdp[2].find_all('a'):
                no_effect.append(a.text.lower())
        except:
            no_effect.append('none')
    else: 
        for a in bmdp[0].find_all('a'):
            not_very_effective.append(a.text.lower())  
        for a in bmdp[1].find_all('a'):
            no_effect.append(a.text.lower())
            
            
    att_lists = [super_effective, not_very_effective, no_effect]
    #list of lists instead of dataframe
    types_dict[type] = att_lists
        
types_dict 
```

```python
def effectiveness(def_pokemon, attack_type, data):
    pokemon_name = def_pokemon
    att_type = attack_type.lower()

    dmg_var = 1

    type1 = str(data[data['name'] == pokemon_name]['type1'].values[0]).lower()
    type2 = str(data[data['name'] == pokemon_name]['type2'].values[0]).lower()
    def_types= [type1,type2]
    

    for type in def_types:
        if type in types_dict[att_type][0]:
            dmg_var *= 2
        if type in types_dict[att_type][1]:
            dmg_var *= .5
        if type in types_dict[att_type][2]:
            dmg_var *=0
    return dmg_var
```

```python
# combine following functions into one func w/ params = super, not_very, no_effect, and standard
def super_rate(type):
    super_counter = 0
    for name in df['name']:
        if effectiveness(name,type,df) > 1:
            super_counter +=1
    super_rate = round(super_counter/len(df['name']),2)*100
    #print(f'{type} is super effective against {super_rate}% of Pokemon')
    return super_rate
```

```python
def not_very_rate(type):
    not_very_counter = 0
    for name in df['name']:
        if effectiveness(name,type,df) < 1:
            not_very_counter +=1
    not_very_rate = round(not_very_counter/len(df['name']),2)*100
    #print(f'{type} is super effective against {super_rate}% of Pokemon')
    return not_very_rate
```

```python
def no_effect_rate(type):
    no_effect_counter = 0
    for name in df['name']:
        if effectiveness(name,type,df) == 0:
            no_effect_counter +=1
    no_effect_rate = round(no_effect_counter/len(df['name']),2)*100
    #print(f'{type} is super effective against {super_rate}% of Pokemon')
    return no_effect_rate
```

```python
def standard_rate(type):
    standard_counter = 0
    for name in df['name']:
        if effectiveness(name,type,df) == 1:
            standard_counter +=1
    standard_rate = round(standard_counter/len(df['name']),2)*100
    #print(f'{type} is super effective against {super_rate}% of Pokemon')
    return standard_rate
```

```python
super_rates = []
not_very_rates = []
no_effect_rates = []
standard_rates = []

for type in types:
    super_rates.append(super_rate(type))
    not_very_rates.append(not_very_rate(type))
    no_effect_rates.append(no_effect_rate(type))
    standard_rates.append(standard_rate(type))
dict = {'Types': types, 'super_rates': super_rates, 'not_very_rates': not_very_rates, 'no_effect_rates': no_effect_rates, 'standard_rates': standard_rates }
pd.DataFrame(dict)
```

```python
types_list = ['Normal',
 'Steel',
 'Dragon',
 'Flying',
 'Rock',
 'Dark',
 'Water',
 'Fairy',
 'Electric',
 'Fighting',
 'Fire',
 'Ground']
weeding_list = [f'{type1},{type2},{type3},{type4}', f'{type1},{type2},{type4},{type3}', f'{type1},{type4},{type3},{type2}',
               f'{type4},{type2},{type3},{type1}',f'{type1},{type3},{type2},{type4}',f'{type3},{type2},{type1},{type4}',f'{type2},{type1},{type3},{type4}']

breakdown_dict = {}
for type1 in types_list:
    for type2 in types_list:
        for type3 in types_list: 
            for type4 in types_list:
                weeding_list = [f'{type1},{type2},{type4},{type3}', f'{type1},{type4},{type3},{type2}',f'{type4},{type2},{type3},{type1}',f'{type1},{type3},{type2},{type4}',f'{type3},{type2},{type1},{type4}',f'{type2},{type1},{type3},{type4}']

                if f'{type1},{type2},{type3},{type4}' not in weeding_list or f'{type1},{type2},{type3},{type4}' not in breakdown_perm.keys(): ## keep a continous breakdown_dict that remembers old calculations and skips them.
                    bd = breakdown(type1,type2,type3,type4)
                    effectiveness_score = bd[2]*2 + bd[3]
                    breakdown_dict[f'{type1},{type2},{type3},{type4}'] = effectiveness_score
breakdown_perm= breakdown_dict
```
