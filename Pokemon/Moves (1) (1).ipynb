```python
def pull_moves(pokemon):
    # Df for moves a pokemon knows
    
    url = f'https://pokemondb.net/pokedex/{pokemon}'
    data = requests.get(url).text
    soup = BeautifulSoup(data,"lxml")
    narrowing1 = soup.body.main.find('div', class_ = 'tabset-moves-game sv-tabs-wrapper').find('div', class_="grid-col span-lg-6")
    narrowing2 = narrowing1.find_all('table', class_ = "data-table")

    all_move_blocks = []
    for block in narrowing2:
        move_blocks = block.tbody.find_all('tr')
        all_move_blocks.append(move_blocks)
    all_move_blocks

    move_ = all_move_blocks[0][1].find_all('td')
    all_move_list =[]


    for i in range(len(all_move_blocks[0])):
        move_ = all_move_blocks[0][i].find_all('td')
        move_info = []
        for j in range(len(move_)):
            if j == 3:
                move_info.append(move_[j]['data-sort-value'].capitalize())
            elif j == 4:
                try:
                    move_info.append(int(move_[j].text))
                except: 
                    move_info.append(int(0))
            else: 
                move_info.append(move_[j].text)
        all_move_list.append(move_info)

    for i in range(len(all_move_blocks[1])):
        move_ = all_move_blocks[1][i].find_all('td')
        move_info = []
        move_info.append('evolve')
        for j in range(len(move_)):
            if j == 2:
                move_info.append(move_[j]['data-sort-value'].capitalize())
            elif j == 3:
                try:
                    move_info.append(int(move_[j].text))
                except: 
                    move_info.append(int(0))
            else: 
                move_info.append(move_[j].text)
        all_move_list.append(move_info)

    for i in range(len(all_move_blocks[2])):
        move_ = all_move_blocks[2][i].find_all('td')
        move_info = []
        move_info.append('egg')
        for j in range(len(move_)):
            if j == 2:
                move_info.append(move_[j]['data-sort-value'].capitalize())
            elif j == 3:
                try:
                    move_info.append(int(move_[j].text))
                except: 
                    move_info.append(int(0))
            else: 
                move_info.append(move_[j].text)
        all_move_list.append(move_info)

    all_move_list   

    pokemon= 'tinkaton'
    url = f'https://pokemondb.net/pokedex/{pokemon}'
    data = requests.get(url).text
    soup = BeautifulSoup(data,"lxml")
    narrowing1 = soup.body.main.find('div', class_ = 'tabset-moves-game sv-tabs-wrapper').find_all('div', class_="grid-col span-lg-6")
    tm_moves = narrowing1[1].tbody.find_all('tr')

    for i in range(len(tm_moves)):
        all_tm_moves= tm_moves[i].find_all('td')
        move_info = []
        for j in range(len(all_tm_moves)):
            if j == 3:
                move_info.append(all_tm_moves[j]['data-sort-value'].capitalize())
            elif j == 4:
                try:
                    move_info.append(int(all_tm_moves[j].text))  
                except: 
                    move_info.append(int(0))
            else:
                move_info.append(all_tm_moves[j].text)
        all_move_list.append(move_info) 



    move_dict2 = {'Level': [],'Name': [], 'Type': [], 'Cat': [], 'Power': [], 'Acc': []}

    move_df2 = pd.DataFrame(move_dict2)

    for move in all_move_list:
        move_df2.loc[len(move_df2.index)] = move
    return move_df2



```
