```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
url2 = https://pokemondb.net/pokedex/game/scarlet-violet
```

```python
df2= pd.read_csv("PokemonTypes.csv")
my_pokemon = df2.name[0:20]
my_pokemon
```

```python
pop = pd.DataFrame(columns=["hp","attack", "defense", "sp_Attack", "sp_Defense", "speed", "total", "name"])

for pokemon in my_pokemon:

    name = pokemon
    url = f'https://pokemondb.net/pokedex/{name}'

    data  = requests.get(url).text
    soup = BeautifulSoup(data,"lxml")

    main_body= soup.body.main
    y= main_body.find('div', class_= "grid-col span-md-12 span-lg-8").tbody
    stats = y.find_all('tr')

    base_stats = []
    for stat in stats:
        value =stat.find('td').text
        base_stats.append(value)

    total = sum(list(map(int, base_stats)))
    base_stats.append(total)
    base_stats.append(name)

    
    pop.loc[len(pop)] = base_stats
pop
```
