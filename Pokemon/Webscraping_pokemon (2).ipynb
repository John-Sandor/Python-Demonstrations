```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
url2 = 'https://pokemondb.net/pokedex/game/scarlet-violet'

data2= requests.get(url2).text
soup2 = BeautifulSoup(data2,"lxml")

```

```python
narrowing = soup2.find("div", class_ = "infocard-list infocard-list-pkmn-lg")
infocards = narrowing.find_all('div', class_="infocard")

pokemon9_list = []
for infocard in infocards:
    pokemon9 = infocard.find("span", class_ = "infocard-lg-data text-muted").a.text
    pokemon9_list.append(pokemon9)
```

```python
df2= pd.read_csv("PokemonTypes.csv")
my_pokemon = df2.name
```

```python
new_pokemon= list(set(pokemon9_list) - set(my_pokemon))
```

```python
 pop = pd.DataFrame(columns=["hp","attack", "defense", "sp_Attack", "sp_Defense", "speed", "total", "name", "type1", "type2"])

def scrap_stats(input_pokemon_list):

   
    for pokemon in input_pokemon_list:

        name = pokemon
        url = f'https://pokemondb.net/pokedex/{name}'

        data  = requests.get(url).text
        soup = BeautifulSoup(data,"lxml")

        body_main= soup.body.main
        
        try: 
            ## Base Stats
            narrowing2= body_main.find('div', class_= "grid-col span-md-12 span-lg-8").tbody
            stats = narrowing2.find_all('tr')

            base_stats = []
            for stat in stats:
                value =stat.find('td').text
                base_stats.append(value)

            ## Total Stats
            total = sum(list(map(int, base_stats)))
            base_stats.append(total)
            
            ## Pokemon Name
            base_stats.append(name)
            
            ## Types
            narrowing3 = soup.body.main
            narrowing4 = narrowing3.find('div', class_= 'sv-tabs-panel-list')
            narrowing5 = narrowing4.find('div', class_= "grid-col span-md-6 span-lg-4").tbody.find_all('tr')

            for a in narrowing5[1].td.find_all('a'):
                base_stats.append(a.text)
            
            if len(base_stats) == 9:
                base_stats.append(' ')
            
            pop.loc[len(pop)] = base_stats
            
            
        except: print(f'well, {name} is missing')


        
    return pop
```

```python
## TESTING
scrap_stats(new_pokemon[0:50])
```
